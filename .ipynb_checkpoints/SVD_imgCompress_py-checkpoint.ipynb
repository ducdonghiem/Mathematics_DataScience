### SVD

```python
A = img_array
```

```python
AtA = A.T @ A
```

```python
AtA.shape
```

```python
eigenvalues, eigenvectors = numpy.linalg.eig(AtA)
len(eigenvalues)
```

```python
eigenvalues[:3]
```

```python
condition_number = numpy.linalg.cond(A)
print("Condition number of A:", condition_number)
```
